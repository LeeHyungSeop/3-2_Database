# getOpenData(endpoing, key) : connection 연결

```python
def getOpenData(endpoint, key) :
    url = endpoint
    params ={
        'serviceKey' : key,
        'page' : 1,
        'perPage' : 1000, # 전체 data는 647개임
        'returnType' : 'JSON'
    }
    res = requests.get(url, params=params).json()
    return res
```
